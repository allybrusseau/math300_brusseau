For this problem I modeled a chained conditional statement and inputed the function we wanted tested. Then told python to print the values for the funtion. I added degenerate triangle because it is also in the description of the problem and i found some cases where the values did equal each other. Technically, a degenerate triangle would not be a triangle but i added it in in order to differenate between the two possiblities of a bad triangle.

```python
def is_triangle(a, b, c):
    if a > (b + c):
        print("No")
    elif b > (a + c):
        print("No")
    elif c > (a + b):
        print("Yes")
    elif a == (b + c):
        print("degenerate triangle")
    elif b == (a + c):
        print("degenerate triangle")
    elif c == (a + b):
        print("degenerate triangle")
```

```python
print(is_triangle(1, 2, 4))
```

```python
print(is_triangle(133, 75, 32))
```

```python
print(is_triangle(1, 2, 3))
```

# 3.) Chapter 6, Problem 2

Defined the function in python given the specifications of 6.5 with recursive functions and modeled it closely to the fibonacci example at the end of 6.7. I included the problem specifications in the code and evaulated it at (3, 4) and then evaulated it at random numbers and found that python did not like that and basically kept running the function! python stated that the maximum recursive depth was exceeded.

```python
def ack(m, n):
    if m == 0:
        return n + 1
    elif m > 0 and n == 0:
        return ack(m - 1, 1)
    elif m > 0 and n > 0:
        return ack(m - 1, ack(m, n -1))
```

```python
print(ack(3, 4))
```

```python
print(ack(5, 9))
```

# 4.) Chapter 6, Problem 5

looking through chapter 5, I modeled this code from the modular example in 5.1 where % is the symbol that takes account for division and the reminder.

```python
def gcd(a,b):
    if b == 0:
        return a
    return gcd(b,a % b)
```

```python
print(gcd(8, 4))
```

```python
print(gcd(130, 12))
```

# 5.) Chapter 10, Problem 2

Defined the function by creating a list. I figured out how to create this code from reading 10.3 and modeled the code from the example in 10.7 that had to do with capitalizing. From there, I plugged in the same numbers as the book and recieved the same answer as the book.

```python
def cumsum(i):
    result = []
    total = 0
    for elem in i:
        total += elem
        result.append(total)
    return result
```

```python
t = [1, 2, 3]
print(cumsum(t))
```
