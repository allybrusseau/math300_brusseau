```python
def lorenz(point, a=10, b=28, c=8/3, dt=0.01):
    (x,y,z) = point
    dx = (a*(y - x))*dt
    dy = (x*(b - z) - y)*dt
    dz = (x*y - c*z)*dt
    return (x+dx, y+dy, z+dz)
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n=1000
point = (1,1,1)
seq = np.empty((n,3))
```

```python
for i in range(n):
    seq[i] = point
    point = lorenz(point)
```

```python
seq
```

```python
plt.plot(seq[:,1], seq[:,2], 'r-')
plt.show()
```
